```python
import torch
```

```python
x = torch.tensor(1.0) # Input
y = torch.tensor(2.0) # Actual value
```

```python
w = torch.tensor(1.0, requires_grad = True) # Weight
```

---

# Forward pass and compute the loss

```python
y_hat = w * x # 1st function ( Big circle thingy)

loss = (y_hat - y)**2 # 2nd function
```

```python
print(loss) 
```

# Backward pass

```python
loss.backward()
print(w.grad)
```
